### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Line Graph 

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.plot(gas.Year, gas.USA, 'b.-', label = 'United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label = 'Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label = 'South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label = 'Australia')


# another way to plot many values
#countries_to_look_at = ['USA', 'Canada', 'Australia', 'South Korea']
#for country in gas:
#    if country in countries_to_look_at:
#        plt.plot(gas.Year, gas[country], marker = '.')

plt.plot(figsize = (8, 5))

plt.xlabel('Year')
plt.ylabel('Price (US dollars)')
plt.title('USA gas price each year', fontdict = {'fontweight': 'bold', 'fontsize' : 18})

print(gas.Year[::3])
plt.xticks(gas.Year[::3].tolist() + [2011])
plt.legend()

plt.savefig('gas_price_figure.png', dpi = 300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
fifa.head(5)
```

### Histograms

```python
bins = [40, 50, 60, 70, 80, 90, 100]
plt.hist(fifa.Overall, bins = bins, color = '#abcdef') #abcdef is the number of the colour
plt.xticks(bins)

plt.ylabel('Numbers of players')
plt.xlabel('Skill level')
plt.title('Distribution of player skills in FIFA 2018') # FIFA 国际足联

plt.show()
```

```python
fifa.head(5)
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.title('Foot Prefernce in FIFA players')
plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f %%')
plt.show()
```

```python
fifa.Weight = [x.strip('lbs') if type(x) == str else x for x in fifa.Weight]
fifa.Weight[0]
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight > 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weight = [light, light_medium, medium, medium_heavy, heavy]
labels =['under 125', '125 - 150', '150 - 175', '175 - 200', 'over 200']
explode = (0.4, 0.2, 0, 0, 0.4)

plt.title('Weight Distribution in FIFA players (in lbs)')
plt.pie(weight, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize =(5, 8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

labels = ['FC Barcelona', 'Real Madrid']

plt.title('Professional Soccer Team Comparison')
boxes = plt.boxplot([barcelona, madrid], labels = labels, patch_artist = True, medianprops = {'linewidth' : 2})

for box in boxes['boxes']:
    # set edge color
    box.set(color = '#abcdef', linewidth = 2)
    # set fill color 
    box.set(facecolor = '#aabbcc')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
